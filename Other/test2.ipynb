```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
data_meta = pd.read_csv(r'test5.csv', encoding='gb2312')
data_meta_array = np.array(data_meta.iloc[0:12, 0:5].values)
```

```python
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_meta_array)
```

```python
tsne = TSNE(n_components=2, perplexity=11)
data_tsne = tsne.fit_transform(data_pca)
```

```python
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], marker='o')
plt.show()
```
