```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

class MemoryBlock:
    def __init__(self, start_address, size, is_free=True, job_name=None):
        self.start_address = start_address
        self.size = size
        self.is_free = is_free
        self.job_name = job_name
    
    def __str__(self):
        status = "空闲" if self.is_free else f"已分配给 {self.job_name}"
        return f"起始地址: {self.start_address}, 大小: {self.size}, 状态: {status}"
```

```python
class MemoryManager:
    def __init__(self, min_size=5):
        self.memory_blocks = []  # 所有内存块列表
        self.last_position = 0   # 用于循环首次适应算法
        self.min_size = min_size # 不可再分割的内存大小
    
    def initialize_from_data(self, data):
        """从输入数据初始化空闲内存块"""
        self.memory_blocks = []
        for start_address, size in data:
            self.memory_blocks.append(MemoryBlock(start_address, size))
        # 按地址排序
        self.memory_blocks.sort(key=lambda block: block.start_address)
    
    def display_memory_status(self):
        """显示当前内存状态"""
        free_blocks = []
        allocated_blocks = []
        
        for block in self.memory_blocks:
            if block.is_free:
                free_blocks.append({
                    '起始地址': block.start_address,
                    '大小': block.size,
                    '状态': '空闲'
                })
            else:
                allocated_blocks.append({
                    '起始地址': block.start_address,
                    '大小': block.size,
                    '状态': '已分配',
                    '作业名': block.job_name
                })
        
        print("空闲内存块：")
        if free_blocks:
            display(pd.DataFrame(free_blocks))
        else:
            print("无空闲内存块")
        
        print("\n已分配内存块：")
        if allocated_blocks:
            display(pd.DataFrame(allocated_blocks))
        else:
            print("无已分配内存块")
        
        self.visualize_memory()
    
    def first_fit(self, size, job_name):
        """首次适应算法"""
        for i, block in enumerate(self.memory_blocks):
            if block.is_free and block.size >= size:
                return self._allocate_memory(i, size, job_name)
        return False
    
    def next_fit(self, size, job_name):
        """循环首次适应算法"""
        # 从last_position开始查找
        for i in range(len(self.memory_blocks)):
            index = (self.last_position + i) % len(self.memory_blocks)
            block = self.memory_blocks[index]
            if block.is_free and block.size >= size:
                self.last_position = (index + 1) % len(self.memory_blocks)
                return self._allocate_memory(index, size, job_name)
        return False
    
    def best_fit(self, size, job_name):
        """最佳适应算法"""
        best_fit_index = -1
        smallest_suitable_size = float('inf')
        
        for i, block in enumerate(self.memory_blocks):
            if block.is_free and block.size >= size:
                if block.size < smallest_suitable_size:
                    smallest_suitable_size = block.size
                    best_fit_index = i
        
        if best_fit_index != -1:
            return self._allocate_memory(best_fit_index, size, job_name)
        return False
    
    def worst_fit(self, size, job_name):
        """最差适应算法"""
        worst_fit_index = -1
        largest_suitable_size = -1
        
        for i, block in enumerate(self.memory_blocks):
            if block.is_free and block.size >= size:
                if block.size > largest_suitable_size:
                    largest_suitable_size = block.size
                    worst_fit_index = i
        
        if worst_fit_index != -1:
            return self._allocate_memory(worst_fit_index, size, job_name)
        return False
    
    def _allocate_memory(self, block_index, size, job_name):
        """分配内存辅助函数"""
        block = self.memory_blocks[block_index]
        
        if block.size - size <= self.min_size:
            # 如果剩余空间小于最小分区大小，则整块分配
            block.is_free = False
            block.job_name = job_name
        else:
            # 分割内存块
            new_block = MemoryBlock(block.start_address + size, block.size - size, True)
            block.size = size
            block.is_free = False
            block.job_name = job_name
            
            # 插入新的空闲块
            self.memory_blocks.insert(block_index + 1, new_block)
        
        return True
    
    def free_memory(self, job_name):
        """释放指定作业的内存"""
        found = False
        for i, block in enumerate(self.memory_blocks):
            if not block.is_free and block.job_name == job_name:
                block.is_free = True
                block.job_name = None
                found = True
                # 合并空闲块
                self._merge_free_blocks()
                break
        return found
    
    def _merge_free_blocks(self):
        """合并相邻的空闲块"""
        i = 0
        while i < len(self.memory_blocks) - 1:
            current = self.memory_blocks[i]
            next_block = self.memory_blocks[i + 1]
            
            if current.is_free and next_block.is_free:
                # 检查是否相邻
                if current.start_address + current.size == next_block.start_address:
                    # 合并两个块
                    current.size += next_block.size
                    self.memory_blocks.pop(i + 1)
                    # 不增加i，因为可能还需要继续合并
                else:
                    i += 1
            else:
                i += 1
    
    def visualize_memory(self):
        """可视化内存使用情况"""
        if not self.memory_blocks:
            print("没有内存块可以显示")
            return
        
        # 计算内存地址范围
        min_addr = min(block.start_address for block in self.memory_blocks)
        max_addr = max(block.start_address + block.size for block in self.memory_blocks)
        
        fig, ax = plt.subplots(figsize=(12, 2))
        
        for block in self.memory_blocks:
            start = block.start_address
            end = start + block.size
            color = 'lightgreen' if block.is_free else 'tomato'
            
            rect = plt.Rectangle((start, 0), block.size, 1, 
                               facecolor=color, edgecolor='black', alpha=0.7)
            ax.add_patch(rect)
            
            # 添加标签
            label = f"空闲 ({block.size})" if block.is_free else f"{block.job_name} ({block.size})"
            plt.text(start + block.size/2, 0.5, label, 
                    ha='center', va='center', fontsize=9)
        
        plt.xlim(min_addr, max_addr)
        plt.ylim(0, 1)
        plt.title('内存使用情况')
        plt.xlabel('内存地址')
        plt.yticks([])
        plt.tight_layout()
        plt.show()
```

```python
# 内存管理系统
manager = MemoryManager(min_size=5)

# 初始化内存空闲区
initial_free_blocks = [
    (0, 100),
    (200, 150),
    (400, 200)
]
manager.initialize_from_data(initial_free_blocks)

print("初始内存状态:")
manager.display_memory_status()

# 测试首次适应算法
print("\n使用首次适应算法分配内存:")
manager.first_fit(50, "Job1")
manager.display_memory_status()

# 测试最佳适应算法
print("\n使用最佳适应算法分配内存:")
manager.best_fit(30, "Job2")
manager.display_memory_status()

# 测试最差适应算法
print("\n使用最差适应算法分配内存:")
manager.worst_fit(80, "Job3")
manager.display_memory_status()

# 测试循环首次适应算法
print("\n使用循环首次适应算法分配内存:")
manager.next_fit(60, "Job3")
manager.display_memory_status()

# 释放内存
print("\n释放Job1的内存:")
manager.free_memory("Job1")
manager.display_memory_status()

# 再次分配
print("\n再次分配内存:")
manager.best_fit(40, "Job4")
manager.display_memory_status()

# 释放Job2的内存
print("\n释放Job2的内存:")
manager.free_memory("Job2")
manager.display_memory_status()
```
