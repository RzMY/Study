```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['font.serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('avgHgt.csv')
x = df['age']
plt.plot(x, df['CHeight'], label='中国男孩身高')
plt.plot(x, df['JHeight'], label='日本男孩身高')
plt.title('中国和日本7-18岁男孩身高图')
plt.xlabel('年龄/岁')
plt.ylabel('身高/厘米')
plt.legend()
plt.show()
```

```python
from PIL import Image, ImageEnhance
img = Image.open('OIP-C.jpg')
plt.imshow(img)
plt.show()
img2 = ImageEnhance.Brightness(img).enhance(2)
plt.imshow(img2)
plt.show()
```
